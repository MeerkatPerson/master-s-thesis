```python
# Runtime is in microseconds, i.e. 0.000001 seconds. So if we divide rt_DEFAULT by 1000000, we get the runtime in seconds.

MICROSECONDS_PER_SECOND = 1000000

PAYLOAD_LEN = 492
CELL_SIZE = 514

# NOTE i realized that in message.py, I was only considering PAYLOAD_LEN when counting data emission. I changed it to CELL_SIZE and am running the experiment again to be sure cuz I'm detail-oriented!!!!!! Verify the cell rate based on this.

# Then just divide the total number of bytes by runtime in seconds, to get bytes per second. Next, determine the number of cells this corresponds to to get the inter-cell interval in seconds. This we can then use as our scale param.
loopix_constant_bidirectional_df["byte_rate_DEFAULT"] = loopix_constant_bidirectional_df["bytes_total_DEFAULT"] / (loopix_constant_bidirectional_df["rt_DEFAULT"] / MICROSECONDS_PER_SECOND)
```

```python
loopix_constant_bidirectional_df["cell_rate_DEFAULT"] =  -(loopix_constant_bidirectional_df["byte_rate_DEFAULT"] // -(CELL_SIZE))
```

```python
# Now for client and server separately? 
loopix_constant_bidirectional_df["byte_rate_client_DEFAULT"] = loopix_constant_bidirectional_df["bytes_client_DEFAULT"] / (loopix_constant_bidirectional_df["rt_DEFAULT"] / MICROSECONDS_PER_SECOND)
# tgen_default_ape["cell_rate_client_DEFAULT"] =  -(tgen_default_ape["byte_rate_client_DEFAULT"] // -(PAYLOAD_LEN))
loopix_constant_bidirectional_df["cell_rate_client_DEFAULT"] =  -(loopix_constant_bidirectional_df["byte_rate_client_DEFAULT"] // -(CELL_SIZE))
```

```python
aggregated_overhead = {'scale': [], 'byte_rate': [], 'cell_rate': [], 'scenario': [], 'overhead_extent': [], 'overhead_std': [], 'overhead_percent': [], 'overhead_type': []}
```

```python
for scenario in ["LOOPIX", "CONSTANT"]:

    for scale_val in [0.0400, 0.0417, 0.0436, 0.0456, 0.0479, 0.0503, 0.0531, 0.0562, 0.0596, 0.0635, 0.0679, 0.0730,
                  0.0789, 0.0859, 0.0942, 0.1043, 0.1169, 0.1328, 0.1538, 0.1827, 0.2250, 0.2927, 0.4186, 0.7347, 3.0]:
        
        scale_val_str = f"{scale_val:.3f}"

        # hack for rounding to three decimals
        scale_val_rounded = float(scale_val_str)

        for overhead_tuple in [("rt", None), ("bytes", "client"), ("bytes", "server")]:

            overhead_type = overhead_tuple[0]

            overhead_perspective = overhead_tuple[1]

            # append the respective scale
            aggregated_overhead['scale'].append(scale_val_rounded)

            # ... and byte rate, which is derived from it (one cell à 514 bytes every 'scale' seconds, thus 1/'scale' * 514 bytes per second)
            byte_rate = 1/scale_val_rounded * 514

            aggregated_overhead['byte_rate'].append(byte_rate)

            aggregated_overhead['cell_rate'].append(-(byte_rate // -514))

            # append the respective scenario
            aggregated_overhead['scenario'].append(scenario)

            # append the overhead type's value and type
            overhead_value = loopix_constant_bidirectional_df[f'{overhead_type}_oh_{overhead_perspective}_{scenario}_(_,_,{scale_val_str})'].median() if overhead_perspective != None else loopix_constant_bidirectional_df[f'{overhead_type}_oh_{scenario}_(_,_,{scale_val_str})'].median()

            # append the overhead type's value and type
            overhead_std = loopix_constant_bidirectional_df[f'{overhead_type}_oh_{overhead_perspective}_{scenario}_(_,_,{scale_val_str})'].std() if overhead_perspective != None else loopix_constant_bidirectional_df[f'{overhead_type}_oh_{scenario}_(_,_,{scale_val_str})'].std()
            
            aggregated_overhead['overhead_extent'].append(overhead_value)

            aggregated_overhead['overhead_std'].append(overhead_std)

            overhead_percent = int((overhead_value - 1) * 100)

            aggregated_overhead['overhead_percent'].append(overhead_percent)

            aggregated_overhead['overhead_type'].append(f"{overhead_type}_{overhead_perspective}")
```

```python
aggregated_overhead_df = pandas.DataFrame.from_dict(aggregated_overhead)
```

```python
aggregated_overhead_loopix = aggregated_overhead_df[aggregated_overhead_df["scenario"] == "LOOPIX"]
```
